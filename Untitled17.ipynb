```python
from datasets import load_dataset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import AutoTokenizer, AutoModelForMaskedLM, Trainer, TrainingArguments
```

```python
# Load dataset
ds = load_dataset("CUTD/sanad_df")
df = pd.DataFrame(ds['train'])
```

```python
df
```

```python
# Encode labels
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['label'])
```

```python
import matplotlib.pyplot as plt
df['label'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
```

#spliting the data

```python
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv02")
model = AutoModelForMaskedLM.from_pretrained("aubmindlab/bert-base-arabertv02")
```

# Preprocess the data


```python
def preprocess_data(sentences):
    inputs = tokenizer(sentences['text'], truncation=True, padding='max_length', max_length=512, return_tensors='pt')
    inputs['labels'] = inputs['input_ids'].clone()

    if isinstance(sentences['label'][0], str):
        unique_labels = set(sentences['label'])
        label_map = {label: i for i, label in enumerate(unique_labels)}
        inputs['label'] = [label_map[l] for l in sentences['label']]
    else:
        inputs['label'] = sentences['label']
    return inputs
```

```python
# Create datasets
tokenized_dataset= ds.map(preprocess_data, batched=True, remove_columns=["Unnamed: 0"])
```

```python
tokenized_dataset
```

```python
# Set training arguments
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="no",
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    num_train_epochs=1
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'],

)

```

```python
# Start training
trainer.train()
```

```python
from transformers import pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model="aubmindlab/bert-base-arabertv02")
```
